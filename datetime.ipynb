```python
import pandas as pd
```

```python
data_orang = pd.read_csv('./dataset/data_orang.csv')
```

```python
data_orang.head()
```

```python
data_orang.info()
```

```python
data_orang["tanggal_lahir"] = pd.to_datetime(data_orang['tanggal_lahir'])
```

```python
data_orang.info()
```

```python
data_orang['Tahun_lahir'] = data_orang['tanggal_lahir'].dt.year
```

```python
data_orang
```

```python
data_orang['Bulan_lahir'] = data_orang['tanggal_lahir'].dt.month           
data_orang['Bulan_nama'] = data_orang['tanggal_lahir'].dt.month_name()     
```

```python
data_orang.head()
```

```python
data_orang['Hari_tanggal'] = data_orang['tanggal_lahir'].dt.day
data_orang['Hari_ke_minggu'] = data_orang['tanggal_lahir'].dt.weekday
data_orang['Hari_nama'] = data_orang['tanggal_lahir'].dt.day_name()
```

```python
data_orang.head()
```

```python
data_orang.to_csv('./data_orang_date.csv', index=False)
```
